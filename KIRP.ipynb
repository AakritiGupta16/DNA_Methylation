```python
import os
import pandas as pd

# Path to the main folder containing subfolders
main_folder_path = 'D:\KIRP_early'

# Create an empty DataFrame to store the data
data_df = pd.DataFrame()

# Maximum number of CpG sites to process per file
max_cpg_sites = 100000

# Iterate through the subfolders
for folder_name in os.listdir(main_folder_path):
    subfolder_dir = os.path.join(main_folder_path, folder_name)
    if os.path.isdir(subfolder_dir):
        for file_name in os.listdir(subfolder_dir):
            if file_name.endswith('.txt'):
                txt_file = os.path.join(subfolder_dir, file_name)
                with open(txt_file, 'r') as file:
                    file_data = {}
                    cpg_count = 0
                    for line in file:
                        parts = line.strip().split('\t')
                        if len(parts) == 2 and parts[0].startswith('cg'):
                            cpg_site, value = parts
                            file_data[cpg_site] = value
                            cpg_count += 1
                            if cpg_count >= max_cpg_sites:
                                break
                    if file_data:
                        # Add the file data to the DataFrame with the filename as the index
                        data_df = data_df.append(file_data, ignore_index=True)

# Write the DataFrame to a CSV file
csv_file = 'D:\kirp_early2.csv'
data_df.to_csv(csv_file, index=False)

print(f'CSV file "{csv_file}" has been created successfully.')
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import auc

# Replace these with the paths to your Excel files
excel_files = [
    r"D:\xgb-kirp-methy-tprs.xlsx",
    "D:\svm-kirp-methy-tprs.xlsx",
    "D:/rf-kirp-methy-tprs.xlsx",
    "D:\knn-kirp-methy-tprs.xlsx",
    "D:/nb-kirp-methy-tprs.xlsx"
]

# Names of the algorithms
algorithm_names = ['XGB','SVM', 'RF', 'KNN', 'NB']

# Initialize lists to store averaged TPRs and AUCs
mean_tprs, aucs = [], []

# Read each Excel file and compute the average TPR
for file_path, name in zip(excel_files, algorithm_names):
    # Read the TPR values from the Excel file
    df = pd.read_excel(file_path, header=None)
    
    # Average the TPR values across all rows (validation sets)
    mean_tpr = df.mean(axis=0).values
    
    # Generate a common FPR array (evenly spaced)
    fpr = np.linspace(0, 1, len(mean_tpr))
    
    # Calculate the AUC
    roc_auc = auc(fpr, mean_tpr)
    
    # Store the averaged TPR and AUC
    mean_tprs.append(mean_tpr)
    aucs.append(roc_auc)

# Plot the ROC curves
plt.figure(figsize=(10, 8))
for i in range(len(mean_tprs)):
    plt.plot(fpr, mean_tprs[i], label=f'{algorithm_names[i]} (AUC = {aucs[i]:.3f})')

# Plot the diagonal line for no-skill classifier
plt.plot([0, 1], [0, 1], linestyle='--', color='gray')

# Label the axes and the graph
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Average ROC Curves for Various Algorithms')
plt.legend(loc='lower right')

# Show the plot
plt.show()
```
